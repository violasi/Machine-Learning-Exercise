# Question 1 矩陣運算

1. 讀取 matrixA.txt 和 matrixB.txt 矩陣
2. 進行矩陣乘法 - matrixA * matrixB， 將得到的矩陣數值，由小到大排序後輸出至ans_one.txt


```python
matrixA = []
for i in open("matrixA.txt"):
    row = [int(x) for x in i.split(",")]
    matrixA.append(row)

matrixB = []
for j in open("matrixB.txt"):
    row = [int(x) for x in j.split(",")]
    matrixB.append(row)

print(matrixA)
print(matrixB)
```

```python
import numpy as np
matrixA = np.array(matrixA)
matrixB = np.array(matrixB)
print(matrixA)
print(matrixB)
```

```python
ans = matrixA.dot(matrixB)
print(ans)
```

```python
ans.sort(axis=1)
print(ans)
np.savetxt("Q1_ans.txt", ans, fmt="%d", delimiter="\r\n")
```
